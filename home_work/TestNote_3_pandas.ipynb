```python
reaction_old_gamers = football[football['Age'] == football['Age'].max()]['Reactions'].mean()
reaction_young_gamers = football[football['Age'] == football['Age'].min()]['Reactions'].mean()
print(round(reaction_old_gamers - reaction_young_gamers, 2))
```

```python
football[football['Value'] > football['Value'].mean()].describe(include='object')
```

```python
wage_max_gkreflex = football[(football['Position'] == 'GK') 
                             & (football['GKReflexes'] == football['GKReflexes'].max())]['Wage'].mean()
wage_max_gkhandling = football[(football['Position'] == 'GK') 
                               & (football['GKHandling'] == football['GKHandling'].max())]['Wage'].mean()

print(round(wage_max_gkreflex / wage_max_gkhandling, 2))
```

```python
shotpower_max_agression = football[football['Aggression'] == football['Aggression'].max()]['ShotPower'].mean()
shotpower_min_agression = football[football['Aggression'] == football['Aggression'].min()]['ShotPower'].mean()

print(round(shotpower_max_agression / shotpower_min_agression, 2))
```
