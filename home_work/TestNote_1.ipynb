```python
import re
pattern = re.compile('\d+')

exams = []

f = open('StudentsPerformance.csv')

for line in f:
    info = line.split(',')
    if info[0] == '"gender"':
        continue
    else:
        new_line = []
        for item in info:
            if pattern.search(item) != None:
                new_line.append(int(pattern.search(item)[0]))
            else:
                new_line.append(item[1:-1])
        exams.append(new_line)
        
f.close()
```

```python
exams
```

```python
sum_reading_score = 0
for row in exams:
    sum_reading_score += row[-2]
    
avrg_reading_score = sum_reading_score / 1000    

students_low_reading_score = 0
for row in exams:
    if row[-2] < avrg_reading_score:
        students_low_reading_score += 1
    
print(students_low_reading_score)
```

```python
sum_female_reading_score = 0
sum_female = 0

for row in exams:
    if row[0] == 'female':
        sum_female_reading_score += row[-2]
        sum_female += 1
    
avrg_female_reading_score = sum_female_reading_score / sum_female
print(avrg_female_reading_score)
```

```python
students_up_writing_score = 0
students_up_and_lunch_writing_score = 0
for row in exams:
    if row[-1] > 90:
        students_up_writing_score += 1
        if row[3] == 'standard':
            students_up_and_lunch_writing_score += 1
print(students_up_and_lunch_writing_score / students_up_writing_score * 100)
```

```python
math_score_list = [row[-3] for row in exams]
max_math_score = max(math_score_list)

sum_reading_score_with_max_math_score = 0
count_student_score_with_max_math_score = 0

for row in exams:
    if row[-3] == max_math_score:
        sum_reading_score_with_max_math_score += row[-2]
        count_student_score_with_max_math_score += 1
        
avrg_reading_score_with_max_math_score = sum_reading_score_with_max_math_score / count_student_score_with_max_math_score
print(avrg_reading_score_with_max_math_score)
```
